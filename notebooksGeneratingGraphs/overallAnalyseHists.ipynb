```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import scipy.stats
import os

from google.colab import drive
drive.mount('/drive')
```

```python
overallAccuracyHists = pd.read_csv('/drive/MyDrive/Colab Notebooks/Thesis Notebooks/results/overallAccuracyHists.csv')
```

```python
allTrainAccs = [[float(j) for j in i[1:-1].split(",")] for i in overallAccuracyHists["trainAccHistory"]]
allValAccs = [[float(j) for j in i[1:-1].split(",")] for i in overallAccuracyHists["valAccHistory"]]
len(allTrainAccs)
i=2
allTrainAccs[i]
overallAccuracyHists["epochs"][i]
```

```python
def graphAcc(i):
    plt.plot(np.arange(overallAccuracyHists["epochs"][i]), allTrainAccs[i], label="Row " + str(i) + ": training accuracy")
    plt.plot(np.arange(overallAccuracyHists["epochs"][i]), allValAccs[i], label="Row " + str(i) + ": validation accuracy")
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title("Training Accuracy History")
    plt.legend(bbox_to_anchor=(1, 0.5, 0.5, 0.5))
    plt.show();

graphAcc(1)
graphAcc(2)
```
