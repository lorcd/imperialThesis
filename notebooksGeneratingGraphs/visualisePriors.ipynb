```python
import tensorflow as tf
import tensorflow_probability as tfp
tfd = tfp.distributions
```

```python
# Additional imports and setting fixed random seed to have reproducibility

import matplotlib.pyplot as plt
import numpy as np
tf.random.set_seed(123)
```

```python
tf.__version__
```

```python
distNormal = tfd.Normal(loc = 0, scale = 1)
distOther = tfd.Normal(loc = 0, scale = 0.5)
# Univariate density functions

import seaborn as sns

t = np.linspace(-8, 8, 10000)
densities1 = distNormal.prob(t) 
densities2 = distOther.prob(t) 

sns.lineplot(t, densities1, label = 'normal')
sns.lineplot(t, densities2, label='other')
plt.ylabel('Probability density')
plt.xlabel('Value')
plt.legend()
plt.show()
```

```python
#horseshoe: potentially good one!!!
"""
The so-called 'horseshoe' distribution is a Cauchy-Normal scale mixture, 
proposed as a sparsity-inducing prior for Bayesian regression. [1] It is 
symmetric around zero, has heavy (Cauchy-like) tails, so that large coefficients 
face relatively little shrinkage, but an infinitely tall spike at 0, which 
pushes small coefficients towards zero. It is parameterized by a positive 
scalar scale parameter: higher values yield a weaker sparsity-inducing effect.
"""
distNormal = tfd.Normal(loc = 0, scale =1)
distOther = tfd.Horseshoe(10)
# Univariate density functions

import seaborn as sns

t = np.linspace(-8, 8, 10000)
densities1 = distNormal.prob(t) 
densities2 = distOther.prob(t) 

sns.lineplot(t, densities1, label = 'normal')
sns.lineplot(t, densities2, label='other')
plt.ylabel('Probability density')
plt.xlabel('Value')
plt.legend()
plt.show()
```

```python
distOther = tfd.Horseshoe(10*tf.ones((2,)))
batch_ndims = tf.size(distOther.batch_shape_tensor())
multiDist = tfd.Independent(distOther, reinterpreted_batch_ndims = batch_ndims)
plt_sample = multiDist.sample(100000)
plt.scatter(plt_sample[:, 0], plt_sample[:, 1], marker='.', alpha=0.05)
plt.axis('equal')
plt.show()
```

```python
#scale mixture normals
distNormal = tfd.Normal(loc = 0, scale = 1)
mix = 0.5
distOther = tfd.Mixture(
  cat=tfd.Categorical(probs=[mix, 1.-mix]),
  components=[
    tfd.Normal(loc=0., scale=np.e**3),
    tfd.Normal(loc=0., scale=np.e**-3),
])
import seaborn as sns

t = np.linspace(-8, 8, 10000)
densities1 = distNormal.prob(t) 
densities2 = distOther.prob(t) 

sns.lineplot(t, densities1, label = 'normal')
sns.lineplot(t, densities2, label='other')
plt.ylabel('Probability density')
plt.xlabel('Value')
plt.legend()
plt.show()
```
